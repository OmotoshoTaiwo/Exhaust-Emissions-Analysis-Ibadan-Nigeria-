```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot
sns.scatterplot(data=pf, x='CO (%)', y='NOx (ppm)', hue='Fuel type')

# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(pf['CO (%)'], pf['NOx (ppm)'])[0, 1]

# Add correlation value to the plot
plt.text(x=pf['CO (%)'].max(), y=pf['NOx (ppm)'].min(), s=f'Correlation: {correlation_coefficient:.2f}', verticalalignment='bottom', horizontalalignment='right')

# Plot line graph for correlation
x = np.linspace(pf['CO (%)'].min(), pf['CO (%)'].max(), 100)
y = correlation_coefficient * x
plt.plot(x, y, color='black', linestyle='--', label='Correlation')

plt.title('CO vs NOx (ppm)')
plt.legend(title='Fuel type')
plt.savefig('CO_vs_Nox_with_correlation.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Remove rows with problematic data points
pf_cleaned = pf.dropna(subset=['CO (%)', 'NOx (ppm)'])

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pf_cleaned, x='CO (%)', y='NOx (ppm)', hue='Fuel type')

# Calculate correlation coefficient for diesel
diesel_data = pf_cleaned[pf_cleaned['Fuel type'] == 'diesel']
diesel_corr_coeff = np.corrcoef(diesel_data['CO (%)'], diesel_data['NOx (ppm)'])[0, 1]

# Calculate correlation coefficient for gas
gas_data = pf_cleaned[pf_cleaned['Fuel type'] == 'gas']
gas_corr_coeff = np.corrcoef(gas_data['CO (%)'], gas_data['NOx (ppm)'])[0, 1]

# Add correlation value for diesel to the plot
plt.text(x=diesel_data['CO (%)'].max(), y=diesel_data['NOx (ppm)'].min(), s=f'Diesel Correlation: {diesel_corr_coeff:.2f}', verticalalignment='bottom', horizontalalignment='right')

# Add correlation value for gas to the plot
plt.text(x=gas_data['CO (%)'].max(), y=gas_data['NOx (ppm)'].min(), s=f'Gas Correlation: {gas_corr_coeff:.2f}', verticalalignment='bottom', horizontalalignment='right')

plt.title('CO vs NOx (ppm)')
plt.xlabel('CO (%)')
plt.ylabel('NOx (ppm)')
plt.legend(title='Fuel type')
plt.savefig('CO_vs_Nox_Diesel_and_Gas.png')
plt.show()
```
